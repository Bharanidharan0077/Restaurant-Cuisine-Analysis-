**Exploratory Data Analysis**

**Ⅰ.1. Top Cuisines**

**Determine the top three most common cuisines in the dataset.**

```python
# Split cuisines into row
cuisines = (df["cuisines"].dropna().str.split(", ").explode())

# Count each cuisine
top_3_cuisines_series = cuisines.value_counts().head(3)
print(top_3_cuisines_series)
```

```python
# Visualization
top_3_cuisines = cuisines.value_counts().head(3)

# Plot pie chart
plt.figure(figsize=(5,5))
plt.pie(
    top_3_cuisines.values,
    labels=top_3_cuisines.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("bright")
)
plt.title("Top 3 Most Common Cuisines")
plt.show()
```

**Calculate the percentage of restaurants that serve each of the top cuisines.**

```python
# Total number of restaurants
total_restaurants = len(df)

top_3_cuisines_percentage = (
    top_3_cuisines / total_restaurants * 100
).round(2)

result = pd.DataFrame({
    "Restaurant_Count": top_3_cuisines,
    "Percentage (%)": top_3_cuisines_percentage
})

print(result)
```

**Calculate the average rating for restaurants in each city.**

```python
#Average Rating Per City
df['aggregate_rating'] = pd.to_numeric(df['aggregate_rating'], errors = 'coerce')

avg_rating_per_city = df.groupby('city')['aggregate_rating'].mean().sort_values(ascending = False)
print(avg_rating_per_city.head())
```

**Determine the city with the highest average rating.**

```python
# Visualization
#Highest Average Rating cities
plt.figure(figsize=(7,5))
x = avg_rating_per_city.head().index
y = avg_rating_per_city.head().values
bars = sns.barplot(x = x, y = y, palette = 'CMRmap')

plt.title("Highest Average Rating Cities", fontsize = 16)
plt.xlabel("City", fontsize = 14)
plt.ylabel("Average Rating", fontsize = 14)

for i, v in enumerate(avg_rating_per_city.head().values):
    plt.text(i, v + 0.01, f"{v:.1f}", ha = "center", va = "bottom", fontsize=11)
plt.show()
    
```

**Ⅰ.3.Price Range Distribution**

**Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.**

```python
#Histogram chart for price range distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["price_range"], discrete=True, edgecolor= "black")
plt.title("Distribution of Price Ranges", fontsize = 16)
plt.xlabel("Price Range", fontsize = 14)
plt.ylabel("Number of Restaurants", fontsize = 14)
plt.xticks(ticks=range(1, 5))  # Assuming price ranges from 1 to 4
plt.tight_layout()
plt.grid(False)
plt.show()
```

**Calculate the percentage of restaurants in each price range category.**

```python
#Percentage of Restaurants in Each Price Range
price_counts = df["price_range"].value_counts(normalize=True) * 100
price_percent = price_counts.sort_index().round(2)

for price, percent in price_percent.items():
    print(f"Price Range {price}: {percent}% of restaurants")
```

**Ⅰ.4.Online Delivery**

**Determine the percentage of restaurants that offer online delivery.**

```python
#Percentage of Restaurants offering Online Delivery
total_restaurants = len(df)
delivery_restaurants = (df['has_online_delivery'] == 'Yes').sum()

percentage_delivery = (delivery_restaurants / total_restaurants) * 100
print(f"{percentage_delivery:.2f}% of restaurants offer online delivery")
```

**Compare the average ratings of restaurants with and without online delivery.**

```python
# Calculate the Percentage of Average Ratings :With vs Without Online Delivery
average_ratings = df.groupby('has_online_delivery')['aggregate_rating'].mean()

print(average_ratings)
```

```python
#Visualization
#Average Ratings: With vs Without Online Delivery
x = average_ratings.index
y = average_ratings.values
plt.figure(figsize=(8,6))
bars = sns.barplot(x=x, y=y, palette = 'CMRmap')
plt.title("Average Ratings: Online Delivery vs No Online Delivery", fontsize = 16)
plt.xlabel("Online Delivery", fontsize = 14)
plt.ylabel("Average Rating", fontsize = 14)
for i, v in enumerate(average_ratings.values):
    plt.text(i, v + 0.02, f"{v:.2f}", ha = "center", va = "bottom", fontsize = 12)
plt.show()
```
